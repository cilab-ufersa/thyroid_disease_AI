```python
import pandas as pd
import joblib
from dalex import Explainer
import matplotlib.pyplot as plt

model = joblib.load('C:\\Users\\caiom\\Desktop\\Sist Hypo\\thyroid_disease_AI\\thyroid_disease_AI\\models_file\\StackingModel.sav')

input_train_path = 'C:\\Users\\caiom\\Desktop\\Sist Hypo\\thyroid_disease_AI\\thyroid_disease_AI\\datasets\\hypothyroid\\input_train.csv'
output_train_path = 'C:\\Users\\caiom\\Desktop\\Sist Hypo\\thyroid_disease_AI\\thyroid_disease_AI\\datasets\\hypothyroid\\output_train.csv'
input_test_path = 'C:\\Users\\caiom\\Desktop\\Sist Hypo\\thyroid_disease_AI\\thyroid_disease_AI\\datasets\\hypothyroid\\input_test.csv'
output_test_path = 'C:\\Users\\caiom\\Desktop\\Sist Hypo\\thyroid_disease_AI\\thyroid_disease_AI\\datasets\\hypothyroid\\output_test.csv'

X_train = pd.read_csv(input_train_path)
# X_test = pd.read_csv(input_test_path)
y_train = pd.read_csv(output_train_path)
# y_test = pd.read_csv(output_test_path, header=None, squeeze=True)

exp = Explainer(model, X_train, y_train)

vip = exp.model_parts()
vip.plot()
plt.savefig('imported_features.png')
```

```python
pdp = exp.model_profile(variables='TSH')
pdp.plot()
# plt.savefig('dependencia_parcial.png')
```

```python
adp = exp.model_profile(type='accumulated', variables='TSH')
adp.plot()
# plt.savefig('dependencia_acumulada.png')
```

```python
bd = exp.predict_parts(X_train.iloc[0, :])
bd.plot()
# plt.savefig('decomposicao_de_predicoes.png')
```

```python
exp.model_diagnostics()
```

```python
'''
O mtodo model_surrogate() do objeto Explainer  utilizado para criar um modelo de surrogado, ou seja, um modelo que  treinado para imitar o comportamento do modelo principal. 
Esse modelo de surrogado pode ser interpretado mais facilmente do que o modelo original, e  frequentemente usado para anlises explicativas.

Ao criar um modelo de surrogado, o objetivo  ter um modelo mais simples que capture as tendncias gerais do modelo principal. Isso  til quando o modelo original  complexo e 
difcil de interpretar.
'''

exp.model_surrogate()
```
