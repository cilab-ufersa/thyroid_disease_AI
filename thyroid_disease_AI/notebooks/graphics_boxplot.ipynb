```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Abrindo dataset
dataset = pd.read_csv('C:\\Users\\caiom\\Desktop\\Sist Hypo\\thyroid_disease_AI\\thyroid_disease_AI\\datasets\\hypothyroid\\hypothyroid_dataset_clean.csv')

dataset.head()
```

```python
for feature in dataset[:1]:
    dataFrame = pd.DataFrame(dataset[feature])
    dataFrame = dataFrame.astype(float)

    # Criar um boxplot para todas as features no DataFrame 'df'
    plt.figure(figsize=(10, 6))  # Tamanho da figura
    plt.boxplot(dataFrame.values, vert=False)  # Cria o boxplot
    plt.xticks(range(1, len(dataFrame.columns) + 1), dataFrame.columns)  # Rótulos no eixo x
    plt.title('Boxplot de Features')  # Título do gráfico

    plt.show()
```
