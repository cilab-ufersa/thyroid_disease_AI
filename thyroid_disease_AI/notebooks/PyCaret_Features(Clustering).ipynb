```python
                # 3 clusters(1, 2, 3, 4)
list_clusters = [['T3', 'TT4', 'T4U measured', 'FTI measured', 'FTI'],
                 ['TSH', 'T3', 'TT4 measured', 'TT4', 'T4U measured', 'T4U', 'FTI measured', 'FTI'],
                 ['on thyroxine', 'TSH', 'T3 measured', 'T3', 'TT4 measured', 'TT4', 'T4U measured', 'T4U', 'FTI measured', 'FTI'],
                 ['on thyroxine', 'query hyperthyroid', 'psych', 'TSH', 'T3 measured', 'T3', 'TT4 measured', 'TT4', 'T4U measured', 'T4U', 'FTI measured', 'FTI'],
                 # 5 clusters(5, 6, 7, 8)
                 ['pregnant', 'psych', 'T3 measured', 'TT4 measured', 'TT4'],
                 ['pregnant', 'psych', 'T3 measured', 'T3', 'TT4 measured', 'TT4', 'T4U', 'FTI'],
                 ['pregnant', 'psych', 'TSH measured', 'TSH', 'T3 measured', 'T3', 'TT4 measured', 'TT4', 'T4U', 'FTI'],
                 ['pregnant', 'query hyperthyroid', 'lithium', 'psych', 'TSH measured', 'TSH', 'T3 measured', 'T3', 'TT4 measured', 'TT4', 'T4U', 'FTI'],
                 # 8 clusters(9, 10, 11, 12)
                 ['I131 treatment', 'T3 measured', 'TT4', 'T4U measured', 'FTI'],
                 ['I131 treatment', 'psych', 'TSH', 'T3 measured', 'T3', 'TT4', 'T4U measured', 'FTI'],
                 ['I131 treatment', 'psych', 'TSH', 'T3 measured', 'T3', 'TT4', 'T4U measured', 'T4U', 'FTI measured', 'FTI'],
                 ['I131 treatment', 'query hyperthyroid', 'psych', 'TSH', 'T3 measured', 'T3', 'TT4 measured', 'TT4', 'T4U measured', 'T4U', 'FTI measured', 'FTI'],
                 # 10 clusters(13, 14, 15, 16)
                 ['sick', 'pregnant', 'T3 measured', 'TT4 measured', 'T4U measured'],
                 ['sick', 'pregnant', 'lithium', 'T3 measured', 'TT4 measured', 'TT4', 'T4U measured', 'FTI'],
                 ['sick', 'pregnant', 'I131 treatment', 'lithium', 'T3 measured', 'T3', 'TT4 measured', 'TT4', 'T4U measured', 'FTI'],
                 ['sick', 'pregnant', 'I131 treatment', 'lithium', 'TSH measured', 'TSH', 'T3 measured', 'T3', 'TT4 measured', 'TT4', 'T4U measured', 'FTI'],
                 # 12 clusters(17, 18, 19, 20)
                 ['sick', 'I131 treatment', 'T3 measured', 'TT4 measured', 'T4U measured'],
                 ['on antithyroid medication', 'sick', 'I131 treatment', 'lithium', 'T3 measured', 'TT4 measured', 'TT4', 'T4U measured'],
                 ['on antithyroid medication', 'sick', 'I131 treatment', 'lithium', 'goitre', 'psych', 'T3 measured', 'TT4 measured', 'TT4', 'T4U measured'],
                 ['on antithyroid medication', 'sick', 'pregnant', 'I131 treatment', 'lithium', 'goitre', 'psych', 'T3 measured', 'TT4 measured', 'TT4', 'T4U measured', 'FTI']]
cont = 0
for features in list_clusters:
    # abrir o dataset
    dataset = pd.read_csv('/content/hypothyroid.csv')
    # transformando os dados em dados categoricos
    for index in dataset.columns.values:
        dataset[index] = dataset[index].astype("category").cat.codes.values

    # Processo de limpeza do dataset (removendo linhas com dados faltantes)
    for i in dataset.columns.values:
        dataset.drop(dataset[dataset[i] == '?'].index, inplace=True)
    dataset = dataset.drop('TBG', axis=1)

    output_label_dataset = dataset['binaryClass']
    dataset = dataset[features]
    # dataset = dataset.drop(['binaryClass'], axis=1)

    # balancear os dados
    sm = SMOTE(random_state=42, k_neighbors=5)
    dataset_balanced, output_classe = sm.fit_resample(dataset, output_label_dataset)

    dataset_b = pd.DataFrame(dataset_balanced)
    classes = pd.DataFrame(output_classe)

    final_dataset = pd.concat([dataset_b, classes], axis=1)

    num_features = len(features)
    if cont <=20:
      cont+=1
    print('\n')
    texto = f'({cont}) - {num_features} features: '
    texto_formatado = f'<h2>{texto}</h2>'
    display(Markdown(texto_formatado))
    print(final_dataset.head())
    print('\n')

    train_size = 0.8
    session = setup(data=final_dataset, target='binaryClass', train_size=train_size, data_split_shuffle=True, normalize = True)
      
    best_model = compare_models()
    text = f'O melhor classificador: '
    text_formatado = f'<h2>{text}</h2>'
    display(Markdown(text_formatado))
    print(best_model)
```
