```python
import pandas as pd
import joblib
from dalex import Explainer

model = joblib.load('C:\\Users\\caiom\\Desktop\\Sist Hypo\\thyroid_disease_AI\\thyroid_disease_AI\\models_file\\StackingModel.sav')

input_train_path = 'C:\\Users\\caiom\\Desktop\\Sist Hypo\\thyroid_disease_AI\\thyroid_disease_AI\\datasets\\hypothyroid\\input_train.csv'
output_train_path = 'C:\\Users\\caiom\\Desktop\\Sist Hypo\\thyroid_disease_AI\\thyroid_disease_AI\\datasets\\hypothyroid\\output_train.csv'
input_test_path = 'C:\\Users\\caiom\\Desktop\\Sist Hypo\\thyroid_disease_AI\\thyroid_disease_AI\\datasets\\hypothyroid\\input_test.csv'
output_test_path = 'C:\\Users\\caiom\\Desktop\\Sist Hypo\\thyroid_disease_AI\\thyroid_disease_AI\\datasets\\hypothyroid\\output_test.csv'

X_train = pd.read_csv(input_train_path)
# X_test = pd.read_csv(input_test_path)
y_train = pd.read_csv(output_train_path)
# y_test = pd.read_csv(output_test_path, header=None, squeeze=True)

exp = Explainer(model, X_train, y_train)
```

```python
exp.model_diagnostics()
```

```python

new_observation = X_train.iloc[0]  # Aqui, utilizamos a primeira observação como exemplo

exp.predict_parts(new_observation) # Visualizando as previsões individuais
```

```python
exp.model_parts()
```

```python
'''
O método model_surrogate() do objeto Explainer é utilizado para criar um modelo de surrogado, ou seja, um modelo que é treinado para imitar o comportamento do modelo principal. 
Esse modelo de surrogado pode ser interpretado mais facilmente do que o modelo original, e é frequentemente usado para análises explicativas.

Ao criar um modelo de surrogado, o objetivo é ter um modelo mais simples que capture as tendências gerais do modelo principal. Isso é útil quando o modelo original é complexo e 
difícil de interpretar.
'''

exp.model_surrogate()
```
